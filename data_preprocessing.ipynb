# Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas import read_csv
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from sklearn.feature_selection import SelectKBest, f_classif
```

# Loading the the raw dataset

```python
dataset_path = '/content/drive/MyDrive/action_recognition/dataset/csh101/csh101.ann.features.csv' 
dataset = read_csv(dataset_path) #this is our raw data
print(f"shape of dataset: {dataset.shape}")
```

#Processing the labels

```python
#converting labels from strings to integers
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(dataset.activity)
dataset['labels'] = encoded_labels #adding a new column: labels
dataset = dataset.drop(['activity'],axis=1) #dropping a column: class names
```

#Standardization

```python
scaler = preprocessing.StandardScaler().fit(dataset.iloc[:,:-1])
dataset.iloc[:,:-1] = scaler.transform(dataset.iloc[:,:-1])
```

```python
X = dataset.iloc[:,:-1] #feature columns
y = dataset.iloc[:,-1]  #target column
```

# Selecting the k most relavent features

```python
selector = SelectKBest(f_classif, k=20)
selector.fit(X, y)
cols = selector.get_support(indices=True) #most important columns chosen by SelectKbest
```

```python
print(selector.get_feature_names_out())
```

```python
dataset = X.iloc[:,cols]  #fetchinf the important columns
dataset['label'] = y
```
